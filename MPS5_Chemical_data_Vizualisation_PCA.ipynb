### Part I: Principal component analysis

In this parctical part, we will familirize with Principal Component Analysis. Principal Component Analysis (PCA) is a method used to make complex data easier to understand. Often, data has too many details that can be confusing. PCA simplifies it by picking out the most important parts, which helps us see the main patterns more clearly. This is especially helpful because we can usually only imagine things in three dimensions, but PCA helps us understand data that goes beyond that.

The key points:
    
    * The method is linear by its nature
    * The main idea of the method is the transformation of the data coordinates onto a new coordinate system such that the directions (principal components) capturing the largest variation in the data
    * You can use the first 2 (for 2D) or 3 (for 3D) principal components for the data visualisation
    
The method is widely used in many practical cheminformatics applications. 

For the good understanding of the method, we recommend you to recall the basis of linear algebra, and, in particular [marix multiplication](https://en.wikipedia.org/wiki/Matrix_multiplication).

##### Intro: Linear transformations. 
Suppose you an objects which can be described by $ K $ numbers. The objects themselves can be of any nature (molecules, spectra, atoms in moleucles). In this case, your data can be described as a vector of length $ K $ 

$
x = \begin{bmatrix}
        x_1 \\
        x_2 \\
        \vdots \\
        x_K
    \end{bmatrix}
$

If you multiply this vector to a matrix $ T $.

$ xT = y $ 

Suppose you have $ N $ objects which can be described by $ K $ numbers. The objects themselves can be of any nature (molecules, spectra, atoms in moleucles). In this case, your data can be described as a matrix $ M $ with $ N $ rows and $ K $ columns. 

$
M_{ij} = \begin{bmatrix}
    a_{11} & a_{12} & \cdots & a_{1K} \\
    a_{21} & a_{22} & \cdots & a_{2K} \\
    \vdots & \vdots & \ddots & \vdots \\
    a_{j1} & a_{j2} & \cdots & a_{NK}
\end{bmatrix}
$



##### PCA machinery 
If we multiply the matrix $ M_{ij} $ with another matrix $ T_{jj} $ we will have the matrix with the same shape, but points will be transformed into a new space. 

$ M^{new} = MT $

But how to find the matrix $ T $?

```python
#SMILES Standardization  
drd2.smiles = drd2.smiles.apply(rdMolStandardize.StandardizeSmiles)
```

```python
from rdkit.Chem import Descriptors
list_of_descriptors = ['MolWt','TPSA','nRotB','HBD','HBA','LogP']
for i in drd2.index:
    mol=Chem.MolFromSmiles(drd2.loc[i,'smiles'])
    drd2.loc[i,'MolWt']=Descriptors.ExactMolWt (mol)
    drd2.loc[i,'TPSA']=Chem.rdMolDescriptors.CalcTPSA(mol) #Topological Polar Surface Area
    drd2.loc[i,'nRotB']=Descriptors.NumRotatableBonds (mol) #Number of rotable bonds
    drd2.loc[i,'HBD']=Descriptors.NumHDonors(mol) #Number of H bond donors
    drd2.loc[i,'HBA']=Descriptors.NumHAcceptors(mol) #Number of H bond acceptors
    drd2.loc[i,'LogP']=Descriptors.MolLogP(mol) #LogP
```

```python
#Normalize the values, by subtracting mean and divide by std (see StandardScaler, )
scaler = StandardScaler()
drd2[list_of_descriptors] = scaler.fit_transform(drd2[list_of_descriptors].values)
```

```python
def visualize_molecules_with_descriptors(df, list_of_descriptors, x_col, y_col, title):
    # Generate RDKit images from DataFrame SMILES
    mols = [Chem.MolFromSmiles(smile) for smile in df['smiles']]
    mol_images = [Draw.MolToImage(mol, size=(300, 150), fitImage=True) for mol in mols]

    # Create figure and plot scatter
    #fig = plt.figure(figsize=(8, 6))
    #ax = fig.add_subplot(111)

    x = df[x_col]
    y = df[y_col]
    plt.ioff()
    fig, ax = plt.subplots()
    plt.ion()
    line, = ax.plot(x, y, ls="", marker="o")
    
    # Set the title and axes labels
    ax.set_title(title, fontsize=16, fontweight='bold', family='sans-serif')
    ax.set_xlabel(x_col, fontsize=14, fontweight='bold')
    ax.set_ylabel(y_col, fontsize=14, fontweight='bold')
    plt.tick_params('both', width=2, labelsize=12)
    plt.tight_layout()
    
    # Create the annotations box with the first image
    zoom_factor = 0.3  # Adjust the zoom factor to scale the images appropriately
    im = OffsetImage(np.array(mol_images[0]), zoom=zoom_factor)
    xybox = (50., 50.)
    ab = AnnotationBbox(im, (0, 0), xybox=xybox, xycoords='data',
                        boxcoords="offset points", pad=0.3, arrowprops=dict(arrowstyle="->"))
    # Add it to the axes and make it invisible
    ax.add_artist(ab)
    ab.set_visible(False)

    def hover(event):
        if line.contains(event)[0]:
            ind, = line.contains(event)[1]["ind"]
            w, h = fig.get_size_inches()*fig.dpi
            ws = (event.x > w/2.)*-1 + (event.x <= w/2.)
            hs = (event.y > h/2.)*-1 + (event.y <= h/2.)
            ab.xybox = (xybox[0]*ws, xybox[1]*hs)
            ab.set_visible(True)
            ab.xy = (x[ind], y[ind])

            # Generate text image with descriptor data
            descriptors_text = '\n'.join([f'{desc}: {df.at[ind, desc]:.2f}' for desc in list_of_descriptors])
            # Create a PIL image for text
            text_height = 200
            text_image = Image.new('RGB', (mol_images[ind].size[0], text_height), (255, 255, 255))

            draw = ImageDraw.Draw(text_image)
            # You may need to adjust the font and size depending on your setup
            # Calculate text position to center it
            text_size = draw.textsize(descriptors_text, font=font)
            text_position = (text_image.size[0] - text_size[0]) // 2, (text_image.size[1] - text_size[1]) // 2
            draw.text(text_position, descriptors_text, fill=(0, 0, 0), font=font)

            # Combine images vertically
            combined_image = Image.new('RGB', (mol_images[ind].size[0], mol_images[ind].size[1] + text_image.size[1]), (255, 255, 255))
            combined_image.paste(mol_images[ind], (0, 0))
            combined_image.paste(text_image, (0, mol_images[ind].size[1]))

            # Set the combined image as the data for the OffsetImage
            im.set_data(combined_image)
        else:
            ab.set_visible(False)
        fig.canvas.draw_idle()

    # Connect the hover function
    cid = fig.canvas.mpl_connect('motion_notify_event', hover)
    plt.show()
```

```python
visualize_molecules_with_descriptors(data,list_of_descriptors,'PC1','PC2',title='Principal Component Analysis, dopamine receptor \n antagonists from DrugBank')
```
